# Seaborn & Pandas Dataframe 

## Statistical Summary 

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

### Missing Value

```python
plt.figure(figsize=(15, 15))
sns.heatmap(-df.isna(), cmap='Blues', cbar=False)
```

### Correlation Heatmap

```python
plt.figure(figsize=(6, 6))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmax=1, annot=True, square=True, fmt=".1f")
```

### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="shelf", vars=["calories", "fat", "rating"])
```

```python
sns.pairplot(df, hue="shelf", vars=["calories", "fat", "rating"], kind=d='reg')
```

```python
sns.jointplot("potass", "fiber", data=df, marginal_kws={"bins": 25}, kind='reg')
```

# Categorical Data 

##  Persebaran

```python
sns.catplot("mfr", "rating", data=df)
```

```python
sns.boxplot("mfr", "rating", data=df)
```

```python
sns.violinplot("mfr", "rating", data=df)
```
