```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialisasi Figure dan Axes 

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

### shortcut 

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r-', linewidth=0.5)
plt.plot(X, y2, 'bo', markersize=3)
plt.xlim(0, 20)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title("Sin dan Cos")
plt.xlabel("X")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set_xlim(0, 20)
ax.set_ylim(-2, 2)
ax.set_title("Sin dan Cos", fontsize=14)
ax.set_xlabel("X", fontsize=10)
ax.set_ylabel("y", fontsize=10)
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set(xlim=(0, 20), ylim=(-2, 2), title="Sin dan Cos", xlabel="X", ylabel="y")
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5, label='sin')
ax.plot(X, y2, 'bo', markersize=3, label='cos')
ax.set(xlim=(0, 20), ylim=(-2, 2), title="Sin dan Cos", xlabel="X", ylabel="y")
ax.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

# Scatter 

```python
df = pd.read_csv("data/circle_data.csv",  delimiter=";")
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X, y)
plt.axis("equal")
```

```python
plt.scatter(X, y, c='r', s=10)
plt.axis("equal");
```

```python
plt.scatter(X, y, c=rad, s=rad*90, cmap='seismic')
plt.axis("equal");
```
