```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialisasi Figure dan Axes 

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

### shortcut 

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r-', linewidth=0.5)
plt.plot(X, y2, 'bo', markersize=3)
plt.xlim(0, 20)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title("Sin dan Cos")
plt.xlabel("X")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set_xlim(0, 20)
ax.set_ylim(-2, 2)
ax.set_title("Sin dan Cos", fontsize=14)
ax.set_xlabel("X", fontsize=10)
ax.set_ylabel("y", fontsize=10)
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set(xlim=(0, 20), ylim=(-2, 2), title="Sin dan Cos", xlabel="X", ylabel="y")
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5, label='sin')
ax.plot(X, y2, 'bo', markersize=3, label='cos')
ax.set(xlim=(0, 20), ylim=(-2, 2), title="Sin dan Cos", xlabel="X", ylabel="y")
ax.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

# Scatter 

```python
df = pd.read_csv("data/circle_data.csv",  delimiter=";")
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X, y)
plt.axis("equal")
```

```python
plt.scatter(X, y, c='r', s=10)
plt.axis("equal");
```

```python
plt.scatter(X, y, c=rad, s=rad*90, cmap='seismic')
plt.axis("equal");
```

# Subplot 

### MATLAB Style

```python
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1);

plt.subplot(233)
plt.hist(numbers, bins=80, density=1);

plt.subplot(235)
plt.scatter(X, y, c='r', s=10)
plt.axis("equal");

plt.subplot(236)
plt.hist(numbers, bins=80, density=1);
```

### Matplotlib Style

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

ax[0, 0].hist(numbers, bins=80, density=1)

ax[0, 1].axis('off')

ax[0, 2].hist(numbers, bins=80, density=1)

ax[1, 0].axis('off')

ax[1, 1].scatter(X, y, c='r', s=10)
ax[1, 1].axis("equal")

ax[1, 2].hist(numbers, bins=80, density=1);
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

ax[0, 0].hist(numbers, bins=80, density=1)

ax[0, 2].hist(numbers, bins=80, density=1)

ax[1, 1].scatter(X, y, c='r', s=10)
ax[1, 1].axis("equal")

ax[1, 2].hist(numbers, bins=80, density=1);
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharey=True)

ax[0, 0].hist(numbers, bins=80, density=1)

ax[0, 2].hist(numbers, bins=80, density=1)

ax[1, 1].scatter(X, y, c='r', s=10)

ax[1, 2].hist(numbers, bins=80, density=1);
```

# Text dan Annotation 

```python
numbers = np.random.randn(100000)
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1)
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha="center")
plt.annotate("mean=0", xy=(0, 0.4), xytext=(1, 0.5), ha="center", arrowprops={'arrowstyle': '-|>'})
```

# 3D plot 

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, antialiased=False, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-10, azim=45)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-10, azim=45)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-10, azim=45)
```
